## Preliminary Wrangling

```python
# import all packages and set plots to be embedded inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

%matplotlib inline
```

```python
# Load the data and check the head
loan = pd.read_csv('prosperLoanData.csv')
loan.head()
```

```python
loan.info()
```

Due to the high amount of features, I am going to subset the data to area of interests. I am going to select only the columns that are related to our exploration of the loan. These columns are related to the amount of loan, dates, rates and other characteristics about the borrowers.

```python
df_loan = loan.loc[:, ['BorrowerRate', 'LenderYield', 'ProsperRating (Alpha)', 'ProsperScore',
                        'ListingCreationDate', 'Term', 'LoanStatus', 'ClosedDate', 'BorrowerAPR',
                         'BorrowerState', 'Occupation', 'EmploymentStatus', 'AmountDelinquent',
                         'MonthlyLoanPayment', 'LP_ServiceFees', 'Investors','ListingCategory (numeric)',
                         'IncomeRange', 'StatedMonthlyIncome','LoanCurrentDaysDelinquent',
                         'LoanOriginalAmount', 'LoanOriginationDate', 'ListingCategory (numeric)',
                         'CreditScoreRangeLower', 'CreditScoreRangeUpper']]
df_loan.head()
```

```python
# check the info
df_loan.info()
```

```python
#description of the numeric columns in dataset
df_loan.describe()
```

```python
df_loan.duplicated().sum()
```

```python
df_loan.isnull().sum()
```

```python
#Change Datetime columns
df_loan['ListingCreationDate'] = pd.to_datetime(df_loan['ListingCreationDate']) 
df_loan['ClosedDate'] = pd.to_datetime(df_loan['ClosedDate'])
df_loan['LoanOriginationDate'] = pd.to_datetime(df_loan['LoanOriginationDate'])
```

```python
# Convert ProsperRating & IncomeRange to categorical 
cat_columns = {'ProsperRating (Alpha)': ['N/A','HR','E','D','C', 'B', 'A', 'AA'],
                    'IncomeRange':['$0','$1-24,999','$25,000-49,999','$50,000-74,999','$75,000-99,999','$100,000+']}

for col in cat_columns:
    cat_col = pd.api.types.CategoricalDtype(ordered = True, categories = cat_columns[col])
    df_loan[col] = df_loan[col].astype(cat_col)
```

```python
# Change dataype for other categorical columns
df_loan['BorrowerState'] = df_loan['BorrowerState'].astype('category')
df_loan['LoanStatus'] = df_loan['LoanStatus'].astype('category')
df_loan['ListingCategory (numeric)'] = df_loan['ListingCategory (numeric)'].astype('category')
df_loan['EmploymentStatus'] = df_loan['EmploymentStatus'].astype('category')
df_loan['Occupation'] = df_loan['Occupation'].astype('category')
```

```python
df_loan.info()
```

## Univariate Exploration

> For this phase, we would be exploring the different variables in the dataset to see how they're distributed, their absolute frequencies and to generally understand their relevance in the entire exploration.

```python
# creating a function for chart name, x and y labels, to avoid repetition in using the plt function
def labels(a,b,c):
    """
    This function gives chart title, x and y axis
    --------
    It aims to avoid repetition by taking in three parameters for the chart title and both the x and y axis labels
    """
    return plt.title(a),plt.ylabel(b),plt.xlabel(c)
```

### Question 1
What is the Borrower APR (Annual Percentage Rate) distribution?

### Visualization

```python
# Distribution of Borrower APR in percentage 
BorrowerAPR_pct = df_loan['BorrowerAPR']*100
bins = np.arange(df_loan['BorrowerAPR'].min()*100, df_loan['BorrowerAPR'].max()*100+2, 2)

plt.figure(figsize=[10, 5])
plt.hist(x = BorrowerAPR_pct, bins = bins)
plt.xticks([5, 10, 15, 20, 25, 30, 35, 40], ['5','10', '15', '20', '25', '30', '35', '40'])
labels('Borrower APR Distribution (%)','Count','Borrower Annual Percentage Rate (%)');
```

### Observation
We can infer that the Borrower APR is normally distrubuted with a peak between 15% and 20% in addition we have some increase in the 35%.

### Question 2
What is the Loan Original Amount distribution?

### Visualization

```python
# Loan Original Amount Distribution
bins = np.arange(df_loan['LoanOriginalAmount'].min(), df_loan['LoanOriginalAmount'].max()+1000, 1000)

plt.figure(figsize=[10, 8])
plt.hist(data = df_loan, x = 'LoanOriginalAmount', bins = bins)
labels('Loan Original Amount Distribution in $','count','Loan Original Amount in $');
```

### Observation
We see that the distribution is right skewed with a long tail so we will transform to a log scale for clearer insight.

```python
# Transform the describe() to a scale of log10
np.log10(df_loan['LoanOriginalAmount'].describe())
```

```python
# Putting Loan Original Amount in a log-scale
bins = 10 ** np.arange(3, 5+0.025, 0.025)

# Generate the x-ticks you want to apply
ticks = [500, 1e3, 2e3,3e3, 5e3, 1e4, 2e4, 3e4, 5e4]

plt.figure(figsize=[10, 5])
plt.hist(data = df_loan, x = 'LoanOriginalAmount', bins = bins)
plt.xscale('log')
labels('Loan Original Amount distribution with Log scale','count','Loan Original Amount ($)');
# Apply x-ticks
plt.xticks(ticks, labels=['500', '1k', '2k', '3k', '5k', '10k', '20k', '30k', '50K']);
```

### Observation
We can see that the highest value for original loan amount is between 3K and 5K. Also I noticed an increase in amount between 10K and 20K.

## Bivariate Exploration

> For the Bivariate Exploration, we would be comparing two variables and trying to see whether there is a correlation between them. 

```python
df_loan.info()
```

```python
# Identify numeric features
num_cols = ['StatedMonthlyIncome', 'LoanOriginalAmount', 'MonthlyLoanPayment','AmountDelinquent',
            'LoanCurrentDaysDelinquent','Term',
            'BorrowerAPR', 'BorrowerRate', 'LenderYield', 'ProsperScore']
```

### Correlation Matrix

```python
plt.figure(figsize = [12, 8])
sb.heatmap(df_loan[num_cols].corr(), annot = True, fmt = '.3f',
           cmap = 'viridis_r', center = 0)
plt.title('Correlation Matrix'); 
```

We can see :
- Strong positive correlations between Lender yield and Borrower Rate with Borrower APR. Loan original amount and Monthly Loan Payment are also in positive correlation. 

- Negative correlation between Prosper Score & Borrower APR, and Prosper Score & Lender yield. Also We see negative relationship between Loan Original Amount and Borrower APR, means large loans have relatively less annual interest rate from the borrower.

### BorrowerRate vs LoanOriginalAmount vs LenderYield

```python
bins_x = np.arange(0, 0.4925+0.1, 0.1)
bins_y = np.arange(1000, 35000+5000,5000)

plt.figure(figsize=(12,8))
plt.hist2d(data=df_loan, x='BorrowerRate', y='LoanOriginalAmount', cmin=0.5,cmap='viridis_r', 
           bins=[bins_x,bins_y], weights=df_loan['LenderYield']);
labels('BorrowRate vs LoanOriginalAmount vs LenderYield','Loan Original Amount','Borrower Rate')
plt.colorbar(label='Lender Yield');
```
