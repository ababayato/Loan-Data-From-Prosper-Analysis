## Investigation Overview

In this investigation, I wanted to look at the factors that affect a loan to be granted, the reason why people take loan also what amount is granted to certain borrower based on their status and occupation. Furthermore can the borrower's income influence the loan amount.


## Dataset Overview

This data set contains 113,937 loans with 81 variables on each loan, including loan amount, borrower rate (or interest rate), current loan status, borrower income, and many others. The data was subseted to 25 data points to aid my analysis.

```python
# import all packages and set plots to be embedded inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

%matplotlib inline

# suppress warnings from final output
import warnings
warnings.simplefilter("ignore")
```

```python
# load in the dataset into a pandas dataframe
df_loan = pd.read_csv('new_prosper_loan.csv')
```

```python
# creating a function for chart name, x and y labels, to avoid repetition in using the plt function
def labels(a,b,c):
    """
    This function gives chart title, x and y axis
    --------
    It aims to avoid repetition by taking in three parameters for the chart title and both the x and y axis labels
    """
    return plt.title(a),plt.ylabel(b),plt.xlabel(c)
```

## Distribution of Loan Original Amount

The origination amount of the loan in the dataset having the highest loan amount between 3k and 5k with an increase between 10k and 20k. Plotted on a logarithmic scale, the distribution of loan original amount takes on a multimodal shape.

```python
# Putting Loan Original Amount in a log-scale
bins = 10 ** np.arange(3, 5+0.025, 0.025)

# Generate the x-ticks you want to apply
ticks = [500, 1e3, 2e3,3e3, 5e3, 1e4, 2e4, 3e4, 5e4]

plt.figure(figsize=[10, 5])
plt.hist(data = df_loan, x = 'LoanOriginalAmount', bins = bins)
plt.xscale('log')
labels('Loan Original Amount distribution with Log scale','count','Loan Original Amount ($)');
# Apply x-ticks
plt.xticks(ticks, labels=['500', '1k', '2k', '3k', '5k', '10k', '20k', '30k', '50K']);
```

## Borrower's Occupation

Plotting the borrower's occupations shows a large number of people specified their occupation to be "others". Thereby depicting professionals, computer programmers and executives as the top 3 occupations of loan borrowers. 

```python
base_color = sb.color_palette()[0]
plt.figure(figsize=(12,10))
order = df_loan.Occupation.value_counts().index
sb.countplot(data=df_loan, y='Occupation', color=base_color, order=order)
labels('Borrower\'s Occupation','Occupation','count');
```

## Employment Status

We can see that though the people who are employed take the most loans while people who are unemployed or retired  take the least loans.


```python
plt.figure(figsize=(12,8))
order = df_loan['EmploymentStatus'].value_counts().index
sb.countplot(data=df_loan, x='EmploymentStatus', color=base_color, order=order)
plt.title('Employment Status')
plt.xlabel('Status')
plt.xticks(rotation='90');
```

## Income Range Distribution
Income range takes on a range of values, from  0 (in dollars) lowest  to about 100k (in dollars) at the highest. Plotted a bimodal and skewed right distribution that shows people with income range of $25,000-49,999 are the ones who are taking loans the most. 

```python
plt.figure(figsize=[10, 5])
sb.countplot(data=df_loan,x='IncomeRange',color=base_color);
plt.title('Borrower Income Range Distribution');
plt.xticks(rotation=90);
```

## Loan Amount by Income Range
We can see how the higher your income range the higher the chances of getting a high loan amount. We see earners of 50k above can get up to 15k loan amount unlike the people who earn lesser.

```python
plt.figure(figsize=(20,5))
order = ['Not displayed','Not employed','$0','$1-24,999','$25,000-49,999','$50,000-74,999','$75,000-99,999','$100,000+']
sb.violinplot(data=df_loan, x='IncomeRange', y='LoanOriginalAmount', color=base_color, order=order)
plt.xlabel('Income Range')
plt.ylabel('Loan Original Amount')
plt.title('Loan Amount by Income Range', fontsize=15)
plt.xticks(rotation=15);
```

## BorrowerRate vs LoanOriginalAmount vs LenderYield
Plotted on a heatmap, we can see that loans of borrower rate of 0.2 to 0.3 yielded more interest rate on the loan than others.

```python
bins_x = np.arange(0, 0.4925+0.1, 0.1)
bins_y = np.arange(1000, 35000+5000,5000)

plt.figure(figsize=(12,8))
plt.hist2d(data=df_loan, x='BorrowerRate', y='LoanOriginalAmount', cmin=0.5,cmap='viridis_r', 
           bins=[bins_x,bins_y], weights=df_loan['LenderYield']);
labels('BorrowRate vs LoanOriginalAmount vs LenderYield','Loan Original Amount','Borrower Rate')
plt.colorbar(label='Lender Yield');
```
